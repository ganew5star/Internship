# 2. In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then scrape all the products available under that product name. Details to be scraped are: "Brand
Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and
“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
url=[]
```

```python
for page in range(0,7):
    product_url=driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in product_url:
        url1=i.get_attribute('href')
        if url1:
            url.append(url1)
        else:
            url.append('-')
        next1=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div[1]/div/span[1]/div[1]/div[67]/div/div/span/a[3]')
        next1.click
```

```python
len(url)
```

```python
Brand=[]
Name=[]
price=[]
exchan=[]
delivery=[]
availability=[]
```

```python
for ds in url:
    driver.get(ds)
    brand_scrap=driver.find_element(By.XPATH,'//div[@class="a-section a-spacing-none"]')
    bran=brand_scrap.text
    if bran:
        Brand.append(bran)
    else:
        Brand.append('-')
        
    name_scrap=driver.find_element(By.XPATH,'//span[@id="productTitle"]')
    nam=name_scrap.text
    if nam:
        Name.append(nam)
    else:
        Name.append('-')
        
    price_scrap=driver.find_element(By.XPATH,'//span[@class="a-price-whole"]')
    prie=price_scrap.text
    if prie:
        price.append(prie)
    else:
        price.append('-')
      
    try:
        exc=driver.find_element(By.XPATH,'//span[@class="a-declarative"]/div/a')#it is in image block unable to track the text
        edc=exc.text
        if edc:
            exchan.append(edc)
        else:
            exchan.append('-')
    except NoSuchElementException as e:
        exchan.append('-')
    try:
        deli_scrap=driver.find_element(By.XPATH,'//div[@id="mir-layout-DELIVERY_BLOCK"]')
        delo=deli_scrap.text
        delivery.append(delo)
    except NoSuchElementException as e:
        delivery.append('-')
    
    avail_scrap=driver.find_element(By.XPATH,'//div[@id="availability"]')
    aval=avail_scrap.text
    if aval:
        availability.append(aval)
    else:
        availability.append('-')
```

```python
exchane=[]
```

```python
df=pd.DataFrame({'Brand':Brand,'Name of the Product':Name,
                'Price':price,
                'Exchange':exchan,
                'Delivery':delivery,
                'Availability':availability,
                'Product_url':url})
```

```python
df.to_csv('guitar_amazon.csv')
```

# 4# Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be scraped

```python
smartphone=[]
```

```python
phone_scrap=driver.find_elements(By.CLASS_NAME,'KzDlHZ')
for j in phone_scrap:
    phone=j.text
    match=re.search(r'(\w.+)([(])',phone)
    if match:
        smart=match.group(1)
        smartphone.append(smart)
```

```python
color=[]
```

```python
color_scrap=driver.find_elements(By.CLASS_NAME,'KzDlHZ')
for k in color_scrap:
    col=k.text
    match=re.search(r'(\w.+)([(])(\w+ \w+)',col)
    if match:
        cols=match.group(3)
        color.append(cols)
```

```python
ram=[]
```

```python
ram_scrap=driver.find_elements(By.CLASS_NAME,'G4BRas')
for l in ram_scrap:
    ra=l.text
    match=re.search(r'(\d.+)([|])',ra)
    if match:
        rem=match.group(1)
        ram.append(rem)
```

```python
rom=[]
```

```python
rom_scrap=driver.find_elements(By.CLASS_NAME,'KzDlHZ')
for m in rom_scrap:
    ro=m.text
    match=re.search(r'(\w.+)([(])(\w+ \w+)(.+)([)])',ro)
    if match:
        rm=match.group(4)
        rom.append(rm)
```

```python
prim_camera=[]
```

```python
prim_camera_scrap=driver.find_elements(By.CLASS_NAME,'G4BRas')
for n in prim_camera_scrap:
    prim_cam=n.text
    match=re.search(r'(.+)([\n]\d+MP.+[|])(.+)',prim_cam)
    if match:
        pri=match.group(2)
        prim_camera.append(pri)
    else:
            prim_camera.append('-')
```
