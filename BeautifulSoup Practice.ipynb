# 1.Write a python program to display all the header tags from wikipedia.org.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
def get_header_tags(url):
    response=requests.get(url)
    if response.status_code==200:
        soup=BeautifulSoup(response.content)
        header_bags=soup.find_all('span',class_='mw-headline')
        header_texts=[tag.text.strip() for tag in header_bags]
        return header_texts
    else:
        print('Failed to retrieve data from the provided URL.')
url="https://en.wikipedia.org/wiki/Main_Page"
header_tags=get_header_tags(url)
if header_tags:
    print('Header Tags:')
    for tag in  header_tags:
        print(tag)
```

# 2.Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
response=requests.get('https://www.imdb.com/list/ls091520106/')
response
```

```python
def imdb(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content)
    movie_name = []
    movie_rate = []
    movie_year = []
        
    for i in soup.find_all('h3', class_="lister-item-header"):
        movie_name_text=i.text.strip()
        movie_name.append(re.split(r'\W+',movie_name_text))
        
    for s in soup.find_all('div', class_="ipl-rating-widget"):
        rating_text = s.text.strip()
        rating = re.search(r'\d+(\.\d+)?', rating_text) 
        if rating:
            movie_rate.append(float(rating.group())) 
        
                    
    for t in soup.find_all('h3', class_="lister-item-header"):
        movie_year_text=t.text.strip()
        movie_year.append(re.search(r'\((\d{4})\)',movie_year_text)[0])
                   
    df = pd.DataFrame({'Movie Name':movie_name,'Movie Rate':movie_rate,'Movie Year':movie_year})
    return df

url ='https://www.imdb.com/list/ls091520106/'
imdb(url)
```

# 3. Write a python program to scrape mentioned details from dineout.co.in : i) Restaurant name ii) Cuisine iii) Location iv) Ratings v) Image URL.

```python
import re
def dineout(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content)
        restaurant_name = []
        cuisine = []
        location = []
        rating = []
        image_url = []
        
        for i in soup.find_all('div', class_='restnt-info cursor'):
            restaurant_name_text=i.text.strip()
            restaurant_name.append(re.split(r'\s*[,]\s*',restaurant_name_text)[0])
        for s in soup.find_all('span', class_='double-line-ellipsis'):
            cuisine_text=s.text.strip()
            cuisine.append(re.split(r'\s*\|\s*', cuisine_text)[-1])
        for t in soup.find_all('div', class_='restnt-loc ellipsis'):
            location.append(t.text.strip())
        for u in soup.find_all('div', class_='restnt-rating rating-4'):
            rating.append(float(u.text.strip()))
        for t in soup.find_all('div', class_='restnt-rating rating-5'):
            rating.append(float(t.text.strip()))  
        for v in soup.find_all('img', class_='no-img'):
            image_url.append(v['data-src'])
        
       
        
        df = pd.DataFrame({
            'Restaurant-Name': restaurant_name,
            'Cuisine': cuisine,
            'Location': location,
            'Rating': rating,
            'Image-Url': image_url
        })
        return df
    else:
        print('Failed to retrieve data')

url = "https://www.dineout.co.in/delhi-restaurants"
dineout(url)
```

# 4.Write s python program to display list of respected former finance minister of India(i.e. Name , Term of office) from https://presidentofindia.nic.in/former-presidents.htm and make data frame.

```python
def former_president(url):
    response=requests.get(url)
    if response.status_code==200:
        soup=BeautifulSoup(response.content)
        presidents=[]
        for i in soup.find_all('div',class_='president-listing'):
            presidents.append(i.text.strip())
        termofoffice=[]
        for s in soup.find_all('h5'):
            termofoffice.append(s.text.strip())
        df=pd.DataFrame({'Presidents':presidents,'Term of Office':termofoffice})
        return df
    else:
        print('Failed to retrieve the data')
url='https://presidentofindia.nic.in/former-presidents'
former_president(url)
```
