```python
import time

def iphone11review(url,total_reviews=100):#function for the loop
    driver=webdriver.Chrome()
    driver.get(url)
    
    review_summary=[]# initialize the review_summary
    full_review=[]    # initialize the full review     
    
    review_sum=driver.find_elements(By.CLASS_NAME,'z9E0IG')# elements path for review summary
    full_sum=driver.find_elements(By.CLASS_NAME,'ZmyHeo')  # elements path for full review
    for j,k in zip(review_sum,full_sum):  # calling the loop to append the list of data for the both column
        review_summary.append(j.text)
        full_review.append(k.text)
        
    reviews_collected=len(review_summary)
    
    while reviews_collected < total_reviews: #while loop for the moving to next pages of the webpages
        try:
            if reviews_collected == 10 :
                next_button=driver.find_element(By.XPATH,'/html/body/div/div/div[3]/div/div/div[2]/div[13]/div/div/nav/a[11]')
            else:
                next_button=driver.find_element(By.XPATH,'/html/body/div/div/div[3]/div/div/div[2]/div[13]/div/div/nav/a[12]')
    
            next_button.click()
            time.sleep(2)
            
            review_sum=driver.find_elements(By.CLASS_NAME,'z9E0IG')
            full_sum=driver.find_elements(By.CLASS_NAME,'ZmyHeo')
            for j,k in zip(review_sum,full_sum):
                review_summary.append(j.text)
                full_review.append(k.text)
                reviews_collected += 1
        except:
            break
    
    review_summary=review_summary[:total_reviews] # assigning the value of 100 to both the columns
    full_review=full_review[:total_reviews]
    
    dh=pd.DataFrame({'Review_Summary':review_summary,'Full_Review':full_review}) # truncate the columns usind DataFrame
    print(dh)
url='https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART'
iphone11review(url,total_reviews=100)# calling the function
```

```python
def quotes1000(url,total=1000):
    driver=webdriver.Chrome()
    driver.get(url)
    
    topquote=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a')
    topquote.click()
    
    quote=[]
    author=[]
    types_quotes=[]
    
    quote_scrap=driver.find_elements(By.CLASS_NAME,"wrap-block") #Scrapping for Quotes
    for i in quote_scrap:
        quo=i.text
        match=re.sub(r'(\n.*)','',quo)
        quote.append(match)
        
    Authors=driver.find_elements(By.CLASS_NAME,'author') #Scrapping for Authors
    for j in Authors:
        author.append(j.text)
    
    
    Types_Quotes=driver.find_elements(By.CLASS_NAME,'tags') #Scrapping for Types of Quotes
    for k in Types_Quotes:
        types_quotes.append(k.text)
        
    amountofquotes=len(author)  #assigning the value of amountofquotes
    
    while amountofquotes < total:
          
        if amountofquotes==100:
            next1=driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[3]/li[12]/a')
        else:
            next1=driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[3]/li[13]/a')
        
        next1.click()
        
        quote_scrap=driver.find_elements(By.CLASS_NAME,"wrap-block")
        for i in quote_scrap:
            quo=i.text
            match=re.sub(r'(\n.*)','',quo)
            quote.append(match)
        
        Authors=driver.find_elements(By.CLASS_NAME,'author')
        for j in Authors:
            author.append(j.text)
    
    
        Types_Quotes=driver.find_elements(By.CLASS_NAME,'tags')
        for k in Types_Quotes:
            types_quotes.append(k.text)
            
            amountofquotes +=1
           
    
    df=pd.DataFrame({'Quotes':quote,'Author':author,'Types of Quotes':types_quotes})   
    print(df)
            
url= 'https://www.azquotes.com/'
quotes1000(url,total=1000)        
```

```python
data=[]
table=driver.find_element(By.TAG_NAME,'tbody')
rows=table.find_elements(By.TAG_NAME,'tr')
for i in rows:
    cols=i.find_elements(By.TAG_NAME,'td')
    if len(cols)==4:
        PMname=cols[0].text
        born_dead=cols[1].text
        term_of_office=cols[2].text
        remarks=cols[3].text
        data.append([PMname,born_dead,term_of_office,remarks])
```
