```python
#clicking the more matches link to produce more result
more_matches=driver.find_element(By.XPATH,'/html/body/section/div/div/div/div/div/div[2]/div[3]/div[2]/div/button')
more_matches.click()
```

```python
series=[]
```

```python
#fetching series or tournment name
try:
    for i in driver.find_elements(By.XPATH,'//h5[@class="match-tournament-name ng-binding"]'):
        series.append(i.text)
except NoSuchElementException:
    series.append('-')
```

```python
place=[]
```

```python
#fetching location the matches conducting
try:
    for i in driver.find_elements(By.XPATH,'//div[@class="match-venue ng-scope"]'):
        place.append(i.text)
except NoSuchElementException:
    place.append('-')
```

```python
Date=[]
```

```python
#fetching scheduled date for each matches
try:
    for i in driver.find_elements(By.XPATH,'//div[@class="match-date-info"]'):
        Dat=i.text
        match=re.search(r'(.*)(.*)',Dat)
        if match:
            da=match.group(1)
            Date.append(da)
except NoSuchElementException:
    Date.append('-')
```

```python
Time=[]
```

```python

#fetching timing for the scheduled date when match going to play
try:
    for i in driver.find_elements(By.XPATH,'//div[@class="match-date-info"]'):
        tim=i.text
        match=re.search(r'(.*)([\n]\d+.*)',tim)
        if match:
            ti=match.group(2)
            Time.append(ti)
except NoSuchElementException:
    Time.append('-')
```

```python
df=pd.DataFrame({'Series':series,
                'Place':place,
                'Date':Date,
                'Time':Time})
```

```python
url='https://www.imdb.com/list/ls512407256/'
driver=webdriver.Chrome()
driver.get(url)
```

```python
tv_url=[]
```

```python
#fetching url for all the series
try:    
    url_scrap=driver.find_elements(By.XPATH,'//a[@class="ipc-title-link-wrapper"]')
    for i in url_scrap:
        if i:
            tv_url.append(i.get_attribute('href'))
        else:
            tv_url.append('-')
except NoSuchElementException:
    tv_url.append('-')
```

```python
len(tv_url)
```

```python
ds=pd.DataFrame({'url':tv_url})
```

```python
#checking all the urls from the link is unique
ds.url.nunique()
```

```python
title=[]
```

```python
#fetching title for all the tv series
title_scrap=driver.find_elements(By.TAG_NAME,'h3')
for i in title_scrap:
    tit=i.text
    match=re.sub(r'(\d+\.)',"",tit)    
    if match:
        title.append(match)
    else:
        title.append('-')
```

```python
title=title[:200]
```

```python
year_span=[]
```

```python
genre=[]
```

```python
runtime=[]
```

```python
rating=[]
```

```python
votes=[]
```

```python
#fetching the yearspan
year_scrap=driver.find_elements(By.XPATH,'//div[@class="sc-b189961a-7 feoqjK dli-title-metadata"]//span[1]')
for i in year_scrap:
    if i:
        year_span.append(i.text)
    else:
        year_span.append('-')
```

```python
len(year_span)
```

```python
#fetching genre

for dm in tv_url:
    driver.get(dm)
    try:   
        gen=[]
        genre_scrap=driver.find_elements(By.XPATH,'//div[@class="ipc-chip-list--baseAlt ipc-chip-list"]//span')
        for i in genre_scrap:
            if i:
                gen.append(i.text)
            else:
                gen.append('-')
        genre.append(gen)
    except NoSuchElementException:
        genre.append('-')
        
        
 #fetching runtime value       
    try:
        runtime_scrap=driver.find_elements(By.XPATH,'//ul[@class="ipc-inline-list ipc-inline-list--show-dividers sc-d8941411-2 cdJsTz baseAlt"]//li[4]')
        if not runtime_scrap:
            runtime.append('-')
            
        else:
            for i in runtime_scrap:    
                try:
                    runtime.append(i.text)
                except StaleElementReferenceException:# using stale exception for not element not present in the path and not exist
                    runtime.append('-')
    except NoSuchElementException:
        runtime.append('-')
```

```python
len(genre),len(runtime)
```

```python
#fetching rating score
rating_scrap=driver.find_elements(By.XPATH,'//span[@class="ipc-rating-star ipc-rating-star--base ipc-rating-star--imdb ratingGroup--imdb-rating"]')
for i in rating_scrap:
    rat=i.text
    match=re.search(r'(\d+\.\d+)',rat)
    if match:
        ra=match.group(1)
        rating.append(ra)
    else:
        rating.append('-')
```

```python
len(rating)
```

```python
#fetching votes for the series
voting_scrap=driver.find_elements(By.XPATH,'//span[@class="ipc-rating-star--voteCount"]')
for i in voting_scrap:
    if i:
        votes.append(i.text)
    else:
        votes.append('-')
```

```python
len(votes)
```

```python
df=pd.DataFrame({'Title':title,
                'Year_Span':year_span,
                 'Genre':genre,
                 'Runtime':runtime,
                 'Rating':rating,
                 'Total_Votes':votes
                })
```

```python
df
```
