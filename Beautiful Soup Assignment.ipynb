# 1) Write a python program to display IMDB’s Top rated 100 Indian movies’ data https://www.imdb.com/list/ls056092300/ (i.e. name, rating, year ofrelease) and make data frame.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
```

```python
import bs4
print(bs4.__version__)
```

```python
def imdb_movies(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content)
    movie_name=[]
    movie_rating=[]
    year_of_release=[]
    
    
    for i in soup.find_all('h3',class_='lister-item-header'):
        movie_name_text=i.text.strip()
        movie_name.append(re.split('\s*[(]\s*',movie_name_text)[0])
    for s in soup.find_all('div',class_="ipl-rating-star small"):
        movie_rating.append(s.text.strip())
    for m in soup.find_all('h3',class_='lister-item-header'):
        movie_year=m.text.strip()
        year_of_release.append(re.split('\s*[()]\s*',movie_year)[1])
        
        
    df=pd.DataFrame({'Movie_Name':movie_name,'Rating':movie_rating, 'Year_of_Release':year_of_release})
    return df
url='https://www.imdb.com/list/ls056092300/'
imdb_movies(url)
```

# 3. Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:
a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.


```python
def teams(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content)
    
    team=[]
    matches=[]
    points=[]
    rating=[]

    for i in soup.find_all('div',class_='si-table-data si-team'):
        team.append(i.text.strip())
    for s in soup.find_all('div',class_='si-table-data si-matches'):
        matches.append(s.text.strip())
    for m in soup.find_all('div',class_='si-table-data si-pts'):
        points.append(m.text.strip())
    for t in soup.find_all('div',class_='si-table-data si-rating'):
        rating.append(t.text.strip())
            
    team_10=team[1:11]
    matches_10=matches[1:11]
    points_10=points[1:11]
    rating_10=rating[1:11]
    df=pd.DataFrame({'Top_10_ODI_Team':team_10,'Matches':matches_10,'Points':points_10,'Rating':rating_10})
    return df

url='https://www.icc-cricket.com/rankings/team-rankings/mens/odi'
teams(url)
```

# 4. Write a python program to scrape details of all the posts from https://www.patreon.com/coreyms .Scrape the heading, date, content and the likes for the video from the link for the youtube video from the post.


```python
def patreon(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html.parser')
    
    header=[]
    date=[]
    content=[]
    likes=[]
    
    for i in soup.find_all('div',class_="sc-bBHxTw iloeMK"):
        header.append(i.text.strip())
    
    for m in soup.find_all('div',class_="sc-lgu5zg-0 dXpjXs"):
        date.append(m.text.strip())
        
    for s in soup.find_all('div',class_="sc-jrQzAO gnnDyW"):
        content.append(s.text.strip())   
    
    for t in soup.find_all('div',class_="sc-jrQzAO knZGTZ"):
        content.append(t.text.strip())   
    
    df=pd.DataFrame({'Header':header,'Date':date,'Content':content,'No. of Likes':likes})
    return df
url='https://www.patreon.com/coreyms'
patreon(url)
```

c)News Link

```python
news_links=[]
for m in soup.find_all('a',class_='LatestNews-headline'):
    news_link=m.get('href')
    if news_link:
        news_links.append(news_link)
    
df2=pd.DataFrame({'News Link':news_links})
print(df2)
```

```python
import re

news_links = []
dates=[]

for m in soup.find_all('a', class_='LatestNews-headline'):
    news_link = m.get('href') 
    if news_link:  
        news_links.append(news_link)  
        date_match = re.search(r'/(\d{4})/(\d{2})/(\d{2})/', news_link)

        if date_match:
            year = date_match.group(1)
            month = date_match.group(2)
            day = date_match.group(3)
            extracted_date = f"{year}-{month}-{day}"
            dates.append(extracted_date)
        else:
            dates.append(None)

df3 = pd.DataFrame({'Dates':dates})

df3

```

```python
def articles(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content)
    
    title=[]
    date=[]
    authors=[]
    
    for i in soup.find_all('h2',class_='h5 article-title'):
        title.append(i.text.strip())
    for m in soup.find_all('p',class_='article-date'):
        date.append(m.text.strip())
    for l in soup.find_all('p',class_='article-authors'):
        authors.append(l.text.strip())
        
    df=pd.DataFrame({'Title':title,'Date':date,'Authors':authors})
    return df
url='https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded-articles/'
articles(url)
```
